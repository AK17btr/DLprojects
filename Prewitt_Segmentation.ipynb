```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#imgae path
path1 = "/content/Tr-gl_0038.jpg"
path2 = "/content/Tr-me_0016.jpg"
path3 = "/content/Tr-no_0015.jpg"
path4 = "/content/Tr-pi_0038.jpg"

# Open the image.
img1 = cv.imread(path1, cv.IMREAD_GRAYSCALE)
img2 = cv.imread(path2, cv.IMREAD_GRAYSCALE)
img3 = cv.imread(path3, cv.IMREAD_GRAYSCALE)
img4 = cv.imread(path4, cv.IMREAD_GRAYSCALE)
```

```python
cv2_imshow(cv.resize(img4, (300, 300)))
```

```python
def PrewittsWdgeDetection(image):
  kernelx = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
  kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

  # Apply Gaussian blur to reduce noise and improve edge detection
  blurred_img = cv.GaussianBlur(image, (5, 5), 0)

  img_prewittX = cv.filter2D(blurred_img, -1, kernelx)
  img_prewittY = cv.filter2D(blurred_img, -1, kernely)

  edges = cv.addWeighted(img_prewittX, 0.5, img_prewittY, 0.5, 0)

  plt.figure(figsize=(10,12))
  plt.subplot(1,4,1), plt.imshow(image,'gray')
  plt.title('Original Image')
  plt.xticks([]), plt.yticks([])
  plt.subplot(1,4,2), plt.imshow(edges,'gray')
  plt.title('Edges')
  plt.xticks([]), plt.yticks([])
  plt.subplot(1,4,3), plt.imshow(img_prewittX,'gray')
  plt.title('Edges in x direction')
  plt.xticks([]), plt.yticks([])
  plt.subplot(1,4,4), plt.imshow(img_prewittY,'gray')
  plt.title('Edges in y direction')
  plt.xticks([]), plt.yticks([])
  plt.show()
```

```python
for i, j in zip([img1, img2, img3, img4],[path1, path2, path3, path4]):
  print(f'image -> {j.split("/")[2]}')
  PrewittsWdgeDetection(i)
  print('\n')
```
